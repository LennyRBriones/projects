## Pegar la ruta del excel en el cdigo

Una vez que ya tienes copiada la ruta de los archivos hay que pegarla en las variables de abajo que se indica

```python
!pip install pandas
import pandas as pd

evento_duplicados = "/content/tes individual que no respeto asientos.xlsx" #Evento del cual quieres saber qu lugares se duplicaron
evento_comparar = "/content/testabono.xlsx" #Evento que deseas usar para comparar con el evento que presenta duplicados

# Dataframe Gen
df1 = pd.read_excel(evento_duplicados)
df2 = pd.read_excel(evento_comparar)

# Gen column to add the duplicated status
df1['Duplicado'] = df1.apply(lambda row: 'duplicado' if any(df2[['ZONA', 'SECCIN', 'ASIENTO']].eq(row[['ZONA', 'SECCIN', 'ASIENTO']]).all(axis=1)) else '', axis=1)

# Show duplicated rows
duplicados = df1[df1['Duplicado'] == 'duplicado']
print("Duplicados encontrados:")
print(duplicados)

# Save the duplicated values in the first file
df1.to_excel(evento_duplicados, index=False)
```
